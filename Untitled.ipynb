```python
def get_RFC_results(df, n_estimators_):
    X= df.iloc[:, 1:]
    y = df.iloc[:, 0]
    X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.30, random_state=42) 
    
    rfc = RandomForestClassifier(n_estimators = n_estimators_, bootstrap = True,
                    max_features = 'sqrt', n_jobs=-1, verbose = 0, random_state=42)

    #model = clf_gradBoost.fit(X_train, Y_train)
    cv_results = cross_val_score(rfc, X, y, cv=10, n_jobs = -1)

    print("Random forest cross validation(10 folds) mean score:")
    print(cv_results.mean())
    
    forest = RandomForestClassifier(n_estimators = n_estimators_, bootstrap = True,
                    max_features = 'sqrt', n_jobs=-1, verbose = 0, random_state=42).fit(X_train, Y_train)

    y_train_pred = forest.predict(X_train)
    y_test_pred = forest.predict(X_test)
    acc_train = accuracy_score(Y_train, y_train_pred)
    acc_test = accuracy_score(Y_test, y_test_pred)
    print("\nAccuracy(train): " + str(acc_train))
    print("Accuracy(test): " + str(acc_test))
    print("Метрики на тестовой выборке")
    print(classification_report(Y_test, y_test_pred))
    
    return cv_results, acc_train, acc_test, forest
```

<h4>Separation on chunks</h4>

```python
#Row reduction будет происходить с датафреймом curr_df
curr_df = balanced_df
```

```python

def split_dataframe(df, chunk_size = 10): 
    chunks = list()
    num_chunks = len(df) // chunk_size + 1
    for i in range(num_chunks):
        chunks.append(df[i*chunk_size:(i+1)*chunk_size])
    return chunks 

# Разбивает датафрейм на отдельные небольшие чанки, находит среднее значение переменных в них, и собирает датафрейм из этих значений
# Симуляция того, что мы анализируем не отдельные предложения, а наборы из нескольких предложений
def get_chunked_dataframes(_chunk_size, df_list):
    chunked_max_df = pd.DataFrame()
    chunked_mean_df = pd.DataFrame()
    chunked_std_df = pd.DataFrame()
    for df in df_list:
        curr_chunked_df = split_dataframe(df, _chunk_size)
        #df_with_mean = pd.DataFrame(columns=["avg_" + x for x in df.columns])
        df_with_max = pd.DataFrame(columns=df.columns)
        df_with_mean = pd.DataFrame(columns=df.columns)
        df_with_std = pd.DataFrame(columns=df.columns)

        #df_with_max = 
        for i in curr_chunked_df:
            #df_with_mean = pd.concat([df_with_mean, pd.DataFrame([i.mean().to_dict()])], ignore_index=True)
            df_with_max = pd.concat([df_with_max, pd.DataFrame([i.max(numeric_only=True)])], ignore_index=True)
            df_with_mean = pd.concat([df_with_mean, pd.DataFrame([i.mean(numeric_only=True)])], ignore_index=True)
            df_with_std = pd.concat([df_with_std, pd.DataFrame([i.std(numeric_only=True)])], ignore_index=True)
            
        chunked_max_df = pd.concat([chunked_max_df, df_with_max], ignore_index=True)
        chunked_mean_df = pd.concat([chunked_mean_df, df_with_mean], ignore_index=True)
        chunked_std_df = pd.concat([chunked_std_df, df_with_std], ignore_index=True)
        
    max_column_names = ["max_" + x for x in chunked_max_df.columns]
    avg_column_names = ["avg_" + x for x in chunked_mean_df.columns]
    std_column_names = ["std_" + x for x in chunked_std_df.columns]
    
    
    chunked_max_df.columns = max_column_names
    chunked_mean_df.columns = avg_column_names
    chunked_std_df.columns = std_column_names
    
    res_chunked_df = chunked_mean_df.join(chunked_max_df)
    res_chunked_df = res_chunked_df.join(chunked_std_df)
    
    return chunked_max_df, chunked_mean_df, chunked_std_df, res_chunked_df


def chunk_dataframe(chunk_size, df):
    grade_grouped_dfs = [y for x, y in df.iloc[:, 0:].groupby('Grade', as_index=False)]
    chunked_max_df, chunked_mean_df, chunked_std_df, chunked_df = get_chunked_dataframes(chunk_size, grade_grouped_dfs)

    chunked_df[chunked_df.columns] = chunked_df[chunked_df.columns].astype(float)

    chunked_max_df[chunked_max_df.columns] = chunked_max_df[chunked_max_df.columns].astype(float)
    chunked_mean_df[chunked_mean_df.columns] = chunked_mean_df[chunked_mean_df.columns].astype(float)
    chunked_std_df[chunked_std_df.columns] = chunked_std_df[chunked_std_df.columns].astype(float)

    chunked_df.drop('max_Grade', inplace=True, axis=1)
    chunked_df.drop('std_Grade', inplace=True, axis=1)
    chunked_df.rename(columns={'avg_Grade':'Grade'}, inplace=True)
    chunked_mean_df.rename(columns={'avg_Grade':'Grade'}, inplace=True)
    chunked_max_df.rename(columns={'max_Grade':'Grade'}, inplace=True)
    chunked_std_df.drop('std_Grade', inplace=True, axis=1)
    chunked_std_df.insert (0, 'Grade', chunked_df.iloc[:, 0])
    
    chunked_df.dropna(inplace = True)
    chunked_mean_df.dropna(inplace = True)
    chunked_max_df.dropna(inplace = True)
    chunked_std_df.dropna(inplace = True)
    return chunked_df, chunked_mean_df, chunked_max_df, chunked_std_df
    
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def get_RFR_results(df, n_estimators_):
    X= df.iloc[:, 1:]
    y = df.iloc[:, 0]
    X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.30, random_state=42) 
    
    rfr = RandomForestRegressor(n_estimators = n_estimators_, bootstrap = True,
                    max_features = 'sqrt', n_jobs=-1, verbose = 0, random_state=42)

    #model = clf_gradBoost.fit(X_train, Y_train)
    cv_results = cross_val_score(rfr, X, y, cv=10, scoring='r2', n_jobs = -1)
    print("Random forest cross validation(10 folds) mean score:")
    print(cv_results.mean())

    forest = RandomForestRegressor(n_estimators = n_estimators_, bootstrap = True,
                    max_features = 'sqrt', n_jobs=-1, verbose = 0, random_state=42).fit(X_train, Y_train)


    y_train_pred = forest.predict(X_train)
    y_test_pred = forest.predict(X_test)
    
    train_mae = mean_absolute_error(Y_train, y_train_pred)
    train_mse = mean_squared_error(Y_train, y_train_pred)
    train_r2 = r2_score(Y_train, y_train_pred)
    
    print(f"Train metrics: \nMAE: {train_mae}\nMSE: {train_mse}\nR2: {train_r2}")
    
    
    test_mae = mean_absolute_error(Y_test, y_test_pred)
    test_mse = mean_squared_error(Y_test, y_test_pred)
    test_r2 = r2_score(Y_test, y_test_pred)
    print(f"Test metrics: \nMAE: {test_mae}\nMSE: {test_mse}\nR2: {test_r2}")

    #print("Метрики на тестовой выборке")
    #print(classification_report(Y_test, y_test_pred))
    
    return cv_results, train_mae, train_mse, train_r2, test_mae, test_mse, test_r2, forest
```

```python
def get_rfr_metrics_based_on_chunk_size(min_, max_, step_, df, n_estimators_):
    x=[]
    cv_results=[]
    train_mae=[]
    train_mse=[]
    train_r2=[]
    test_mae=[]
    test_mse=[]
    test_r2=[]
    for i in range(min_, max_, step_):
        print("\nchunk_size = " + str(i))
        x.append(i)
        if (i == 1):
            curr_chunked_df = df
        else:    
            curr_chunked_df, curr_chunked_mean_df, curr_chunked_max_df, curr_chunked_std_df = chunk_dataframe(i, df)

        curr_cv_results, curr_train_mae, curr_train_mse, curr_train_r2, curr_test_mae, curr_test_mse, curr_test_r2, curr_forest = get_RFR_results(curr_chunked_df, n_estimators_)
        
        cv_results.append(curr_cv_results.mean())
        train_mae.append(curr_train_mae)
        train_mse.append(curr_train_mse)
        train_r2.append(curr_train_r2)
        test_mae.append(curr_test_mae)
        test_mse.append(curr_test_mse)
        test_r2.append(curr_test_r2)
        

    return cv_results, train_mae, train_mse, train_r2, test_mae, test_mse, test_r2, x
```

```python
enum=[10, 50, 100, 200, 300]
cv_results=[]
train_mae=[]
train_mse=[]
train_r2=[]
test_mae=[]
test_mse=[]
test_r2=[]
x=[]
for n in enum:
    curr_cv_results, curr_train_mae, curr_train_mse, curr_train_r2, curr_test_mae, curr_test_mse, curr_test_r2, curr_x = get_rfr_metrics_based_on_chunk_size(1, 100, 5, balanced_df, n)
    cv_results.append(curr_cv_results)
    train_mae.append(curr_train_mae)
    train_mse.append(curr_train_mse)
    train_r2.append(curr_train_r2)
    test_mae.append(curr_test_mae)
    test_mse.append(curr_test_mse)
    test_r2.append(curr_test_r2)
    x.append(curr_x)
```

```python
from sklearn.base import clone 

def drop_col_feat_imp(model, X_train, y_train, random_state = 42):
    
    # clone the model to have the exact same specification as the one initially trained
    model_clone = clone(model)
    # set random_state for comparability
    model_clone.random_state = random_state
    # training and scoring the benchmark model
    model_clone.fit(X_train, y_train)
    benchmark_score = model_clone.score(X_train, y_train)
    # list for storing feature importances
    importances = []
    #print("benchmark_score = " + str(benchmark_score))
    # iterating over all columns and storing feature importance (difference between benchmark and new model)
    for col in X_train.columns:
        model_clone = clone(model)
        model_clone.random_state = random_state
        model_clone.fit(X_train.drop(col, axis = 1), y_train)
        drop_col_score = model_clone.score(X_train.drop(col, axis = 1), y_train)
        #print(col + "drop_col_score = " + str(drop_col_score))
        importances.append(benchmark_score - drop_col_score)
    
    importances_df = imp_df(X_train.columns, importances)
    return importances_df, importances
```
